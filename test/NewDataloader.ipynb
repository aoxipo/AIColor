```python
import os
import glob
file_path =r'E:\Dataset\training_set\train\labels'
file = glob.glob(file_path+'/*.txt')
# print(file)
maxNum = 0

for i in file:
    label = []
    with open(i) as f:
        lines = f.readlines()
        for line in lines:
            x = line.replace('\n', '').split(' ')
            x = [float(i) for i in x]
            label.append(x)
    if(len(label)>maxNum):
        maxNum = len(label)
        print(i)

print(maxNum)
```
